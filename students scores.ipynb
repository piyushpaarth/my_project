```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("student result.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df=df.drop("Unnamed: 0",axis=1)
print(df.head())
```

```python
print(df.head())
```

```python
#gender distribution
plt.figure(figsize=(4,4))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("gender distribution")             
plt.show()
```

```python
#from the above chart,the number of female is more than male.And there is not much difference b/w females and males.
```

```python
gb=df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
plt.figure(figsize= (3,3))
sns.heatmap(gb,annot=True)
plt.title("relation b/w parentedu and student marks")
plt.show()
```

```python
#from above chart,parents with education have good effect on marks.
```

```python
hb=df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(hb)
```

```python
plt.figure(figsize= (3,3))
sns.heatmap(hb,annot=True)
plt.title("relation b/w parent_maritial_status and student scores")
plt.show()
```

```python
#from above graph we concluded that there is not much affect of parent marital status on students scores.
```

```python
groupA=df.loc[(df['EthnicGroup']=="group A")].count()
groupB=df.loc[(df['EthnicGroup']=="group B")].count()
groupC=df.loc[(df['EthnicGroup']=="group C")].count()
groupD=df.loc[(df['EthnicGroup']=="group D")].count()
groupE=df.loc[(df['EthnicGroup']=="group E")].count()
l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
print(mlist)
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title("distribution of ethnic groups")
plt.show()
```
